```python
#~~The CSV file from the CDC is located at https://zzbcnsux.labs.coursera.org/notebooks/assignments/assignment2/assets/NISPUF17.csv
#~~If you don't have access to Coursera, the CSV comes from a .DAT file hosted here: https://ftp.cdc.gov/pub/Vaccines_NIS/NISPUF17.DAT
#~~I downloaded the CSV locally for analysis rather than using requests to pull from online.
import pandas as pd
import numpy as np
df = pd.read_csv('NISPUF17.csv')

#~~Find the proportion of mother's education for all children in the dataset

dct = dict()
def proportion_of_education():
    conditions = [(df['EDUC1']==1),(df['EDUC1']==2),(df['EDUC1']==3),(df['EDUC1']==4)]
    values = ['less than high school','high school','more than high school but not college','college']
    df['EDUCATION']=np.select(conditions,values)
    dflen=len(df)
    for value in values:
        dct[value]=len(df[df['EDUCATION']==value])/dflen
    return dct

    # your code goes here
    # YOUR CODE HERE
#     pass
    raise NotImplementedError()
proportion_of_education()
```

```python
#~~Determine how many influenza vaccines are received by children who we know were breast-fed against those who we know were not breastfed.
df = pd.read_csv('NISPUF17.csv')
df=df[(df['CBF_01'] == 1)|(df['CBF_01'] == 2)]
values = [1,2]
keep=['P_NUMFLU','CBF_01']
df=df[keep]
# df.head()
def average_influenza_doses():
    df1 = df[df['CBF_01']==1]
    df2 = df[df['CBF_01']==2]
    #~~this can also be done using df.loc["SOME CONDITION","COLUMNS WANTED"]
    return(df1['P_NUMFLU'].mean(),df2['P_NUMFLU'].mean())
    raise NotImplementedError()
average_influenza_doses()
```

```python
#~~Determine if the chickenpox vaccine might be less effective for one sex than the other.
#~~In other words, are boys who receive the chickenpox vaccine more likely to catch chickenpox than girls who receive it?
df = pd.read_csv('NISPUF17.csv')
indexes = ['SEX','HAD_CPOX','P_NUMVRC']
df=df.loc[(df['HAD_CPOX'].lt(3))&(df['P_NUMVRC'].gt(0)),indexes]
sexcond = [df['SEX']==1,df['SEX']==2]
sexval = ['male','female']
df['SEX']=np.select(sexcond,sexval)
dct2 = dict()
def chickenpox_by_sex():
    # YOUR CODE HERE
    for val in sexval:
        dct2[val]=len(df.loc[(df['SEX']==val)&(df['HAD_CPOX']==1)])/len(df.loc[(df['SEX']==val)&(df['HAD_CPOX']==2)])
    return dct2
    raise NotImplementedError()
chickenpox_by_sex()
```

```python
#~~Find the correlation coefficient for the odds of getting chickenpox against the amount of chickenpox vaccines received
df = pd.read_csv('NISPUF17.csv')
indexes = ['SEX','HAD_CPOX','P_NUMVRC']
df=df.loc[(df['HAD_CPOX'].lt(3)),indexes].dropna()
# df.head(20)
def corr_chickenpox():
    import scipy.stats as stats
    import numpy as np
    import pandas as pd
    
    corr, pval=stats.pearsonr(df["HAD_CPOX"],df["P_NUMVRC"])
    return corr

    raise NotImplementedError()
corr_chickenpox()
```
